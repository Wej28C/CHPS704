6) Camembert

```python
import matplotlib.pyplot as plt
```

```python
pclass_stat = titanic['pclass'].value_counts()
print(pclass_stat)
```

```python
plt.figure(figsize=(12, 6))
# survived
plt.subplot(1, 2, 1)
plt.pie(survived_stat, labels=survived_stat.index, autopct='%1.1f%%', startangle=90, colors=['lightcoral', 'lightskyblue'])
plt.title('Distribution de survived')

# pclass
plt.subplot(1, 2, 2)
plt.pie(pclass_stat, labels=pclass_stat.index, autopct='%1.1f%%', startangle=90, colors=['red', 'lightgreen', 'lightblue'])
plt.title('Distribution de Pclass')

# Affichage des graphiques
plt.tight_layout()
plt.show()
```

10) Données catégorisées

```python
bins = [0, 12, 18, 60, 100]
labels = ['enfants', 'jeunes', 'adultes', 'seniors']


titanic['age_group'] = pd.cut(titanic['age'], bins=bins, labels=labels, right=False)


print(titanic)

# Distribution des groupes d'âge
age_group_counts = titanic['age_group'].value_counts()

# Création du graphique camembert pour la répartition des groupes d'âge
plt.figure(figsize=(8, 6))
plt.pie(age_group_counts, labels=age_group_counts.index, autopct='%1.1f%%', startangle=90, colors=['lightcoral', 'lightblue', 'lightgreen', 'gold'])
plt.title('Répartition des classes d\'âge')
plt.show()
```

11) Machine Learning


```python
from sklearn.model_selection import train_test_split
```

```python
intercept= pd.get_dummies( titanic, columns=['pclass', 'sex', 'embarked'], drop_first=True)
```

```python
print("Données après encodage:")
print(intercept)
```

```python
titaX = intercept.drop(intercept.columns[-1],axis=1)
titaY = intercept[intercept.columns[-1]]
X_train, X_test, Y_train, Y_test = train_test_split(titaX, titaY,test_size=0.25)
```

```python
print("\nTaille de X_train:", X_train.shape)
print("Taille de X_test:", X_test.shape)
print("Taille de Y_train:", Y_train.shape)
print("Taille de Y_test:", Y_test.shape)
```
